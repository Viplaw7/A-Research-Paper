````python
import numpy as np
import matplotlib.pyplot as plt

# Function to calculate power based on temperature
def calculate_power(temperature):
    # Example power calculation (replace with actual model)
    power = np.array([1.83, 1.825, 1.8125, 1.78, 1.775, 1.77])
    return power

# Function to simulate efficiency based on temperature
def simulate_efficiency(temperature, power_values):
    # Example simulation logic (replace with actual model)
    efficiency = 10 * power_values
    return efficiency

# Set new starting and ending points for temperature values
start_temperature = 25
end_temperature = 50
temperature_values = np.linspace(start_temperature, end_temperature, 6)

# Calculate power values for each temperature
power_values = calculate_power(temperature_values)

# Simulate efficiency for each temperature value
efficiency_values = np.vectorize(simulate_efficiency)(temperature_values, power_values)

# Create a figure with a double y-axis
fig, ax1 = plt.subplots(figsize=(9, 6))

# Plot efficiency on the left y-axis
color = 'tab:red'
ax1.set_xlabel('Temperature (Â°C)')
ax1.set_ylabel('Efficiency', color=color)
ax1.plot(temperature_values, efficiency_values, marker='o', linestyle='-', color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a second y-axis for power on the right
ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Power', color=color)

# Set a specific y-axis limit for the power plot (adjust as needed)
ax2.set_ylim(1.75, 1.85)

ax2.plot(temperature_values, power_values, marker='s', linestyle='--', color=color)
ax2.tick_params(axis='y', labelcolor=color)

# Set title
plt.title('InGaN Solar Cell Efficiency and Power vs Temperature')
````
